# **ML Project**

## **Load Data**

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

# **Data Preparation**

## Data Separation as x and y

```python
y=df["logS"]
```

```python
y
```

```python
x=df.drop("logS" , axis=1)
x
```

## Data Splittng

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

# **Model Building**

## **Linear Regression**  

### Training the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

### Applying the model to a prediction

```python
y_lr_train_pred =lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### Evaluate the model performance


```python
#y_train
```

```python
#y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR_MSE (Train): ',lr_train_mse)
print('LR_R2 (Train): ',lr_train_r2)
print('LR_MSE (Test): ',lr_test_mse)
print('LR_R2 (Test): ',lr_test_r2)
```

```python
lr_results = pd.DataFrame(["Linear Regression" ,lr_train_mse ,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
```

```python
lr_results.columns=["Method","Training_MSE","Training_R2","Testing_MSE","Testing_R2"]
```

```python
lr_results
```

# Random Forest  

## Training the model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

## Appllying the model to make a prediction

```python
y_rf_train_pred =rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

## Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

rf_results = pd.DataFrame(["Random Forest " ,lr_train_mse ,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
rf_results.columns=["Method","Training_MSE","Training_R2","Testing_MSE","Testing_R2"]
```

```python
rf_results
```

# Model comparison

```python
df_models = pd.concat([lr_results,rf_results],axis =0)
df_models
```

```python
df_models.reset_index(drop=True)
```
